# 各縣市老年人口現況分析

#  (圖四) 各縣市男女老年人口數長條圖

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
ep = pd.read_csv("elderly population.csv") #讀取檔案
sns.set(style="darkgrid") #設定圖表的背景顏色
x=ep["District"]
plt.figure(1,figsize=(20,10)) #設定圖表大小
plt.xticks(fontsize=20) #設定x軸標籤字體大小
plt.yticks(fontsize=20) #設定y軸標籤字體大小
y1 = list(ep["male"])
y2 = list(ep["female"])
#設定兩條bar的寬度、顏色、對齊方式、圖例名稱
plt.bar(x,y1,width = 0.4,color = "salmon", align = "edge",label = "male")
plt.bar(x,y2,width = -0.4,color = "lightblue", align = "edge",label = "female")
plt.xticks(rotation=90) #設定兩軸標籤的傾斜角
plt.title("gender analysis of elderly population",fontsize=30) # 加上圖表標題並設定字體大小
plt.xlabel("Counties and Cities",fontsize=30) #加上 x軸標題並設定字體大小
plt.ylabel("Elderly Population",fontsize=30) #加上 y軸標題並設定字體大小
plt.legend(loc='upper right', prop={'size': 20}) #設定圖例的位置與字體大小
plt.show()
```
