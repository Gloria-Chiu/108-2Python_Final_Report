# 各縣市社福移工現況分析

# (圖五) 各縣市社福移工數長條圖

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
mw = pd.read_csv("migrant workers.csv",encoding ="big5")  #讀取檔案
sns.set(style="darkgrid") #設定圖表的背景顏色
plt.figure(1,figsize=(30,10)) #設定圖表大小
plt.xticks(fontsize=20)  #設定x軸標籤字體大小
plt.yticks(fontsize=20)  #設定y軸標籤字體大小
x = list(mw["District"])
y = list(mw["workers"])
position =["salmon","salmon","salmon","salmon","salmon","salmon","lightblue","lightblue","lightblue","lightblue","lightblue","lightblue","lightblue","lightblue","lightblue","lightblue","lightblue","lightblue","lightblue","lightblue","lightblue","lightblue"] #依照位置順序設定長條圖的顏色
plt.bar(x,y,color = position) #繪製出長條圖並加上顏色區別
plt.xticks(rotation=90)  #設定 x軸標籤的傾斜角
plt.legend()
plt.title("Migrant Workers in Taiwan",fontsize=50)  # 加上圖表標題並設定字體大小
plt.xlabel("Counties and Cities",fontsize=30) #加上 x軸標題並設定字體大小
plt.ylabel("Institutions numbers",fontsize=30) #加上 y軸標題並設定字體大小
plt.show()
```
