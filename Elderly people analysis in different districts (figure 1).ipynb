# 各縣市老年人口現況分析

#  (圖一) 各縣市老年人口比例(老年人口 / 總人數)

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df=pd.read_csv("elderly population.csv") #讀取檔案
sns.set(style="darkgrid") #設定圖表的背景顏色
x=list(df['District']) #令 x軸 為檔案中 District欄位的資料
y=list(df['% of total elderly population in the county']) #令 y軸 為檔案中 % of total elderly population in the county欄位的資料
position =["salmon","salmon","salmon","salmon","salmon","salmon","lightblue","lightblue","lightblue","lightblue","lightblue","lightblue","lightblue","lightblue","lightblue","lightblue","lightblue","lightblue","lightblue","lightblue","lightblue","lightblue"] #依照位置順序設定長條圖的顏色
plt.figure(1,figsize=(20,10)) #設定圖表大小
plt.bar(x, y, color=position) #繪製出長條圖並加上顏色區別
plt.title("Comparison of the elderly population", fontsize=30) # 加上圖表標題並設定字體大小
plt.xlabel("City", fontsize=20) #加上 x軸標題並設定字體大小
plt.ylabel("Percentage of the elderly population(%)", fontsize=20) #加上 y軸標題並設定字體大小
plt.xticks(rotation=90, fontsize=20) #設定 x軸標籤的傾斜角和字體大小
plt.show()
```
