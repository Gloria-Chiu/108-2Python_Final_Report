# 各縣市長照機構現況分析

#  圖(七)  各縣市長照機構數量長條圖

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
csvdata = pd.read_csv("service.csv") #讀取檔案
sns.set(style="darkgrid") #設定圖表的背景顏色
plt.figure(1,figsize=(20,10)) #設定圖表大小
plt.xticks(fontsize=20) #設定x軸標籤字體大小
plt.yticks(fontsize=20) #設定y軸標籤字體大小
width=0.3 #設定bar的寬度皆為0.3
x1=csvdata['地區']
y1=csvdata['養護型機構可供進住']
x2=csvdata['地區']
y2=csvdata['長期照護機構可供進住']
x3=csvdata['地區']
y3=csvdata['安養機構可供進住']
x1 = np.arange(len(y1))
#設定三條bar的寬度、顏色、對齊方式、圖例名稱
plt.bar(x1,y1,align = "edge",width = 0.3,label='Respite Care',color = "lightblue")
plt.bar(x1+0.3,y2,align = "edge",width = 0.3,label='Long-term care',color = "salmon")
plt.bar(x1+0.6,y3,align = "edge",width = 0.3,label='Nursing home',color = "teal")
plt.xticks(x1+0.3,list(csvdata['地區']),rotation=90)   #設定 x軸資料範圍和標籤的傾斜角
plt.title("Number of three types of institutions for the elderly population in counties and cities",fontsize=30)  # 加上圖表標題並設定字體大小
plt.legend(loc="upper right",fontsize=20)
plt.xlabel("Counties and Cities",fontsize=30) #加上 x軸標題並設定字體大小
plt.ylabel("Institutions numbers",fontsize=30) #加上 y軸標題並設定字體大小
plt.show()
```
